```python
import matplotlib.pyplot as plt
from sklearn import linear_model
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv ("C:/Users/DANIELA/Desktop/C H E/UNAD/ANALI_DE_DAT/TAREA 2/data.csv",sep=",")
```

```python
data
```

```python
#Linea para grafico de dispersion
data.plot.scatter(x="metro",y="precio")
plt.show()
```

```python
regresion = linear_model.LinearRegression()
```

```python
metro = data ["metro"].values.reshape((-1,1))
```

```python
modelo = regresion.fit(metro,data["precio"])

print("Interseccion(b)", modelo.intercept_)
print("Pendiente (m)", modelo.coef_)
```

```python
entrada = [[5],[15],[20]]
predicciones = modelo.predict(entrada)
print(predicciones)
```

```python
data.plot.scatter(x="metro",y="precio", label='Datos Originales')
plt.scatter(entrada,predicciones, color='green')
plt.plot(entrada, predicciones, color='yellow', label='Linea de regresion')
plt.xlabel('metro')
plt.ylabel('precio')
plt.legend()
plt.show()
```
